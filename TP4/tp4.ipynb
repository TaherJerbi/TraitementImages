```python
from math import floor
import matplotlib.pyplot as plt

import numpy as np
from numpy import array, int32
import random
def show_img(image):
    plt.imshow(image)
    plt.show()
def im_read(path):
    with open(path, 'rb') as pgmf:
        im = plt.imread(pgmf)
    return im

chat = im_read('./chat.ppm')
pepper = im_read('./peppers.ppm')

```

```python



def seuiller(img,r,g,b):
    new_image = np.zeros(img.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            # r
            if img[i][j][0] > r:
                new_image[i][j][0] = 255
            else:
                new_image[i][j][0] = 0
            # g
            if img[i][j][1] > g:
                new_image[i][j][1] = 255
            else:
                new_image[i][j][1] = 0

            #b
            if img[i][j][2] > b:
                new_image[i][j][2] = 255
            else:
                new_image[i][j][2] = 0

    return new_image
chat_seuiller = seuiller(chat,50,100,50)
pepper_seuiller = seuiller(pepper,255,255,0)
show_img(chat_seuiller)
show_img(pepper_seuiller)
```

```python
def seuiller_ET(img, r, g, b):
    new_image = np.zeros(img.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            # r
            if img[i][j][0] > r and img[i][j][1] > g and img[i][j][2] > b:
                new_image[i][j][0] = 255
                new_image[i][j][1] = 255
                new_image[i][j][2] = 255
            else:
                new_image[i][j][0] = 0
                new_image[i][j][1] = 0
                new_image[i][j][2] = 0
            
    return new_image
chat_seuiller_ET = seuiller_ET(chat, 100,50,50)
show_img(chat_seuiller_ET)
pepper_seuiller_ET = seuiller_ET(pepper, 100,50,50)
show_img(pepper_seuiller_ET)
```

```python
def seuiller_OU(img, r, g, b):
    new_image = np.zeros(img.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            # r
            if img[i][j][0] > r or img[i][j][1] > g or img[i][j][2] > b:
                new_image[i][j][0] = 255
                new_image[i][j][1] = 255
                new_image[i][j][2] = 255
            else:
                new_image[i][j][0] = 0
                new_image[i][j][1] = 0
                new_image[i][j][2] = 0
            
    return new_image

chat_seuiller_OU = seuiller_OU(chat, 100,50,50)
show_img(chat_seuiller_OU)
pepper_seuiller_OU = seuiller_OU(pepper, 100,50,50)
show_img(pepper_seuiller_OU)
```
